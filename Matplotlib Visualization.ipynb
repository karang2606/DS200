### Load Necessary Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Basic Graph

```python
df = pd.read_excel('NFHS.xls')
df
```

```python
df.describe()
```

```python
df.columns
```

```python
plt.figure(figsize=(20,7))
plt.bar(df['State/UT'],df['Female population age 6 years and above who ever attended school (%)'])
plt.xticks(fontsize=14, rotation = 90)
plt.yticks(fontsize=14)
plt.xlabel('State/UT', fontsize=16)
plt.ylabel('% Feamle population', fontsize=16)
plt.title('Female population age 6 years and above who ever attended school (%) vs State/UI', fontsize=20)
plt.show()
```

```python
df3=df[['Female population age 6 years and above who ever attended school (%)',
       'Population below age 15 years (%)',
       'Children under age 5 years whose birth was registered with the civil authority (%)',
       'Men age 15 years and above wih Mildly elevated blood pressure (Systolic 140-159 mm of Hg and/or Diastolic 90-99 mm of Hg) (%)',
       'Men age 15 years and above wih Moderately or severely elevated blood pressure (Systolic ≥160 mm of Hg and/or Diastolic ≥100 mm of Hg) (%)',
       'Men age 15 years and above wih Elevated blood pressure (Systolic ≥140 mm of Hg and/or Diastolic ≥90 mm of Hg) or taking medicine to control blood pressure (%)',
       'Women age 15 years and above who use any kind of tobacco (%)',
       'Men age 15 years and above who use any kind of tobacco (%)',
       'Men age 15 years and above who consume alcohol (%)']]
```

```python
df3.columns
df3.info()
```

```python
plt.figure(figsize=(10,10))
plt.boxplot(df3,notch=True,  # notch shape
                         vert=True,  # vertical box alignment
                         patch_artist=True)  # fill with color)
# plt.xticks(fontsize=14, rotation = 90)
# plt.yticks(fontsize=14)
plt.xlabel('Columns of Data', fontsize=16)
plt.ylabel('% population', fontsize=16)
# plt.title('Female population age 6 years and above who ever attended school (%) vs State/UI', fontsize=20)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(25,7))
plt.scatter(df['Number of Women age 15-49 years interviewed'],df['Female population age 6 years and above who ever attended school (%)'])
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
plt.xlabel('Number of Women age 15-49 years interviewed', fontsize=20)
plt.ylabel('% Feamle population', fontsize=20)
plt.title('Female population age 6 years and above who ever attended school (%) vs Number of Women age 15-49 years interviewed', fontsize=22)
plt.show()
```
